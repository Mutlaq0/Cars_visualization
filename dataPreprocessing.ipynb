```python
import pandas as pd
```

```python
df= pd.read_csv('C:\\xampp\\htdocs\\HW2_207500448\\cars_cleaned.csv')
df
```

```python
category =[]
for index, row in df.iterrows():
    if row['SportsCar'] == 1 :
        category.append('SportsCar')
    elif row['SUV'] ==1:
        category.append('SUV')
    else:
        category.append('Other')
category
```

```python
df['category'] = category
df = df[['Cyl','HP','HwyMPG','category']]
df
```

```python
df.to_csv('C:\\xampp\\htdocs\\HW2_207500448\\cars.csv',index=False)
```
